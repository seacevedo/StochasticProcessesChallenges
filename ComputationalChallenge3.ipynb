```
import numpy as np
import matplotlib.pyplot as plt
from numba import jit
import seaborn as sns
sns.set(style='ticks', palette='Set2')
```

```
def run_simulation(lamb, n, MAX_TIME, K, seasonal, max_steps):
  time_points = []
  pop_sizes = [] 
  current_time = 0
  steps = 0

  for i in range(max_steps):
    if seasonal:
      rate = lamb * n + 2 * np.sin(current_time)
    else:
      rate = lamb * n

    current_time += np.random.gamma(K, 1/rate)
    n += K

    time_points.append(current_time)
    pop_sizes.append(n)

  return time_points, pop_s`izes
```

```
time, pop_size = run_simulation(0.4, 1, 100, 2, False, 1000)
time_season, pop_size_season = run_simulation(0.4, 1, 100, 2, True, 1000)
```

```
plt.plot(time, pop_size, label='v(t) = 0.4n')
plt.plot(time_season, pop_size_season, label='v(t) = 0.4n + 2 * sin(t)')
plt.xlabel('Time')
plt.ylabel('Population Size')
plt.legend()
```
